```python
import os
import shutil
from sklearn.model_selection import train_test_split

dataset_location = dataset.location
train_images_dir = os.path.join(dataset_location, "train", "images")
train_labels_dir = os.path.join(dataset_location, "train", "labels")

images = [f for f in os.listdir(train_images_dir) if f.endswith(('.jpg', '.png', '.jpeg'))]

train_images_list, val_images_list = train_test_split(images, test_size=0.2, random_state=42)

val_images_dir = os.path.join(dataset_location, "val", "images")
val_labels_dir = os.path.join(dataset_location, "val", "labels")

os.makedirs(val_images_dir, exist_ok=True)
os.makedirs(val_labels_dir, exist_ok=True)

# Correctly move files to the validation directory
for image_file in val_images_list:  # Iterate only through the validation images
    shutil.move(os.path.join(train_images_dir, image_file), os.path.join(val_images_dir, image_file))  # Use shutil.move
    label_file = image_file.replace(image_file.split(".")[-1], "txt")
    if os.path.exists(os.path.join(train_labels_dir, label_file)):
        shutil.move(os.path.join(train_labels_dir, label_file), os.path.join(val_labels_dir, label_file))

import yaml

data_yaml_path = os.path.join(dataset_location, "data.yaml")

with open(data_yaml_path, 'r') as f:
    data = yaml.safe_load(f)

data['train'] = os.path.join(dataset_location, "train", "images")
data['val'] = os.path.join(dataset_location, "val", "images")
# data['test'] = os.path.join(dataset_location, "test", "images") # If you have a test set

with open(data_yaml_path, 'w') as f:
    yaml.dump(data, f)
```
